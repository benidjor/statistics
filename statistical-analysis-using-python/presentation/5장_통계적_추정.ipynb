# 통계적 추정

## 1. 점추정

<용어 정리>
1. 점추정 : 모집단의 모수를 측정할 때, 표본으로부터 얻은 통계량으로 추정하는 것
2. 추정량 : 추정의 목적으로 사용하는 특정한 통계량
3. 추정치 : 표본으로부터 계산된 추정량의 구체적인 값
즉, 추정량은 일종의 개념이고 추정치는 실제 수치를 나타낸다.

<바람직한 추정량>
1. 불편성 : 추정량의 기댓값이 추정하고자 하는 모수일 경우, 이 값을 불편 추정량이라고 부름. (예: 평균)
2. 효율성 : 두 개의 추정량이 있을 때 더 작은 분산을 갖는 경우 
이 두 가지를 모두 만족하는 추정량이 바람직한 추정량이며, 이를 최소분산불편추정량이라고 부른다.
단, 중앙값은 일반적으로 불편성을 만족한다고 보기 어렵다.
이를 실제 추정에 적용해보면, 모평균에는 표본평균이 가장 좋은 추정치라고 할 수 있다. 이 때, 표본평균의 표준편차를 표준오차라고 한다. 
표준편차랑 헷갈리지 않아야 하는데, 표준편차는 하나의 표본 내에서 확률변수 X에 대한 편차이지만, 표본오차는 표본평균 X에 대한 편차를 나타낸다.

비율의 경우에도, 모집단의 가장 좋은 추정량은 표본비율로 정의할 수 있다.

번외) 일치성과 충분성
1. 일치성 : 표본의 크기가 무한히 증가할 때, 추정량이 모수에 근접하려는 특성. 즉, 큰 수의 법칙을 따른다는 개념
2. 충분성 : 추정량은 모수에 관한 모든 정보를 포함해야 한다. (예를 들어, 평균은 모든 정보를 포함하지만 median은 그렇지 않음

## 2. 구간추정

<배경>
점추정으로는 정확도가 너무 낮을 수 있다는 한계가 존재. 따라서 구간을 예측하여 정확도를 높이고자 하는 목적

<정의>
구간추정이란 신뢰수준이 일정하게 주어지고 모수가 들어있는 구간을 찾는 것. 
예를 들어, 95%의 신뢰수준을 주고, 갑 후보에 대한 지지율이 45%에서 48%의 구간 사이에 있을 것이라고 예측한다면, 이때 신뢰구간은 45% ~ 48%가 된다. 

<신뢰수준의 필요성>
신뢰수준은 지나치게 구간의 범위를 넓혀 추정의 의미를 상실하는 것을 방지하기 위해 필요하다. 

<유의수준 a>
유의수준 a는 오차를 허용하는 수준으로, 신뢰수준은 1-a로 정의될 수 있다. 유의수준에 대한 자세한 설명은 6장에서 진행

## 3. 모평균에 대한 신뢰구간 추정

<모분산 사용 신뢰구간 추정>

우리는 앞서, 표본평균이 모평균에 대한 최소분산불편추정량이라는 것을 확인한 바 있다. 그렇다면 이제 표본평균을 사용하여 모평균에 대한 신뢰구간을 구해보자.

먼저, 표본평균을 표준화하면
$$Z=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$$
로 나타낼 수 있다.

이때, 표준화를 해주는 이유는 표본의 크기가 충분히 크다면 표본평균의 분포가 정규분포로 수렴하기 떄문 (중심극한정리)
따라서 표준화된 상태에서 Z에 대한 신뢰구간을 구하고, 여기에 앞서 표준화하기 전 표본평균(X바)에 대한 식으로 다시 풀어주면

$$\bar{X}-z_{\alpha/2}\frac{\sigma}{\sqrt{n}}, \bar{X}+z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$$

로 정리할 수 있다.


<표본분산 사용 신뢰구간 추정>
그러나 위에서 제시한 방법은 모분산을 알아야 사용가능하다. 하지만 대부분의 경우에 모분산은 알 수 없다. 따라서 모분산 대신 표본분산을 사용하며, 정규분포 Z 대신 자유도가 n-1을 따르는 T분포를 사용하게 된다. 이 경우, 신뢰구간은 다음과 같이 나타낼 수 있다.
$$\bar{X}-t_{\alpha/2}\frac{S}{\sqrt{n}}, \bar{X}+t_{\alpha/2}\frac{S}{\sqrt{n}}$$

왜 t분포를 사용하는가? t분포의 경우 소표본에 더 유리하기 때문

## 4. 모비율에 대한 신뢰구간 추정

표본비율(p_hat) 역시 표본의 크기가 충분히 클 때 평균이 p이고 표준편차가 p(1-P)/n인 정규분포를 따른다고 알려져 있다. 이를 활용하여 (1-a)100% 신뢰수준의 신뢰구간은 모평균 추정 떄와 마찬가지로 표준화를 시키고 이를 표본비율 p_hat에 대해 풀어내어 구할 수 있다. 

$$\hat{p}-z_{\alpha/2}\sqrt{\frac{p(1-p)}{n}}, \hat{p}+z_{\alpha/2}\sqrt{\frac{p(1-p)}{n}}$$

그러나, 모평균 때와 마찬가지로 모비율 p를 알 수 없기 때문에 이를 표본비율로 대체하여 다음과 같이 나타낸다.

$$\hat{p}-z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}, \hat{p}+z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

그러나 모비율 추정 때에는 t분포를 사용할 수 없는데 왜냐면 표본비율 p_hat의 분산이 p(1-p)/n을 따른 다는 전제 자체가 표본의 크기가 커야만 성립하기 때문이다. 따라서 모비율 추정 시에는 t분포를 적용할 수 없다. 
이는 모집단의 정규성을 가정할 수 있다면 30개 이하의 소표본에 t분포를 적용할 수 있는 모평균 추정과는 확연히 다른 점으로, 주의가 필요하다. 

## 5. 모분산에 대한 신뢰구간 추정

이제 모분산에 대한 신뢰구간을 측정해보자. 앞서 분산에 대한 가장 좋은 추정량 (최소분산불편추정량)은 표본분산으로 다음과 같이 표기할 수 있다.

$$S^{2} = \frac{\sum(X_{i}-\bar{X})^2}{(n-1)}$$

이 부분 교재가 뭔ㄱ나 이상한데 엄밀히 말해 표본분산은 최소분산불편추정량이 아니라 불편추정량이다. 왜 n-1로 나누는가 의문이 들 수 있는데, 일반적으로 n으로 나누는 것보다 n-1이 더 불편성을 만족한다고 알려져 있다. (표본의 수가 모집단보다 작기 때문에 표본평균이 과대 대표되는 경향이 있어서 n-1로 나누는 것 정도로 알고 있음)

분산 추정에는 자유도가 n-1인 카이제곱 분포를 사용한다.

$$\frac{(n-1)S^2}{\sigma^{2}} ~ \chi^{2}_{(n-1)}$$

마찬가지로 이를 모분산에 대한 신뢰구간으로 풀어주면 다음과 같이 구할 수 있다.

$$\frac{(n-1)S^2}{\chi^{2}_{\alpha/2}}, \frac{(n-1)S^2}{\chi^{2}_{1-\alpha/2}}$$

이는 다음과 같은 카이제곱 그래프에서 양쪽 꼬리 부분을 제외한 부분을 가리킨다.



참고로, 카이제곱 분포 자체가 확률변수의 제곱값의 분포이므로 (확률변수 - 평균)의 제곱인 분산의 분포를 다루기에 적합하다고 할 수 있다.

## 6. 서로 독립인 두 집단의 평균들 차이에 대한 신뢰구간 추정

서로 독립인 두 집단의 평균 차이에 대한 최소분산불편추정량은 표본평균X - 표본평균Y가 될 것이다. 
따라서 X-Y는 다음과 같은 분포를 따른다.
단, 이 단원의 서로 독립인 두 집단은 엄밀히 말하면 서로 동일한 모분산을 가진 두 집단에 해당한다. 

$$\bar{X} - \bar{Y} ~ N(\mu_{X} - \mu_{Y}, \frac{\sigma^{2}_{X}}{n1}+\frac{\sigma^{2}_{Y}}{n2})$$

분산이 이렇게 나오는 이유가 헷갈릴 수 있는데, 3.7절을 참고하면
var(X_bar - y_bar) = var(X_bar) + var(y_bar) - 2cov(x,y)로 정의할 수 있고, 서로 독립이라 공분산 = 0이므로 x모분산/n1 + y모분산/n2로 정의할 수 있다.

이제 X-Y를 Z로 표준화시켜서 정리한 다음, 알지 못하는 모분산 대신 표분분산 S를 적용한다.
그러나 두 집단의 모분산이 동일하다고 가정하고 S를 통합하여 추정하여 

(X평균 표준편차 제곱합 + Y평균 표준편차 제곱합) / (n1 - 1) + (n2 - 1)

로 통합 분산 추정량을 구할 수 있고, 그 자유도는 n1+n2-2이다. 
이를 마찬가지로 표준화하면 t분포를 따르므로, 이를 정리하면 최종적으로 다음과 같은 신뢰구간을 구할 수 있다.

$$(\bar{X} - \bar{Y}) - t_{\alpha/2}\sqrt{\frac{S^{2}_{X}}{n1}+\frac{S^{2}_{Y}}{n2}}, (\bar{X} - \bar{Y}) + t_{\alpha/2}\sqrt{\frac{S^{2}_{X}}{n1}+\frac{S^{2}_{Y}}{n2}}$$

이때, 자유도는 굉장히 복잡한 값이 나오므로 최대한 가까운 정수로 근사한 값을 사용한다.

## 7. 짝으로 이루어진 자료의 차에 대한 신뢰구간 추정

이번에는 서로 독립인 두 집단이 아닌 하나의 모집단을 대상으로 두 개의 자료를 얻는 경우, 그 평균의 차이를 다룬다. 이는 보통 실험 전/후를 비교하는 데 용이하게 사용된다. 

먼저, 서로 대응되는 확률변수의 차(이를 D라고 표시한다)를 구하고 단일 표본의 모평균 추정 시 사용한 방법과 같은 방법을 사용하면 다음과 같이 신뢰구간을 구할 수 있다.

$$\bar{D}-t_{\alpha/2}\frac{S_{D}}{\sqrt{n}}, \bar{D}+t_{\alpha/2}\frac{S_{D}}{\sqrt{n}}$$

## 8. 표본 크기의 결정

표본크기는 크면 클수록 좋지만 그만큼 비용도 많이 들기 때문에 적절하게 결정해야 한다. 

모집단의 크기가 N일 경우, 표본비율 p_hat의 분산은 이론적으로 정확하게 p(1-p)/n의 값에 (N-n)/(N-1)을 곱한 값을 따른다고 알려져 있다.
이 정보를 사용하여 신뢰 구간을 구한 다음 신뢰구간의 폭(오차 한계 = 허용 오차의 범위 = 표본오차)을 구하면 다음과 같다. (n은 표본 수)

$$z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}\sqrt{\frac{N-n}{N-1}}$$

이때, n의 크기를 여유있게 되도록 큰 값으로 하기 위해 표본 비율을 0.5로 설정하고 n 값을 구할 수 있다. 
단, 이는 N이 유한일 경우에 이론적으로 얻어지는 값으로, 만일 모집단의 크기가 무한히 크다면 앞서 모비율에서 구한 것과 같은 식의 신뢰구간이 나온다. 이는 앞서 모비율 추정에 t분포를 사용하지 못한 이유와 같다. 

모비율이 아니라 모평균 추정 시에도 동일하게 신뢰구간의 폭과 신뢰수준을 이용해서 n값을 구한다. (단, 오차한계의 범위를 지정해줘야 하고, 분산의 크기를 알고 있어야 한다.)

### 용어 정리

1. 표준편차 : 확률변수(X)에 대한 분산의 제곱근
2. 표준오차 : 표본평균(X_bar)에 대한 분산의 제곱근
3. 표본오차(=오차한계) : 모집단을 대표할 수 있는 전형적인 구성 요소를 선택하지 못해 발생하는 오차를 의미하며 임계값(Z나 t값)과 표준오차의 곱(= 신뢰구간의 폭)으로 구할 수 있다. 
4. 신뢰수준 : 표본의 신뢰구간이 모수를 의미하는 비율. 즉 95%의 신뢰수준은 100개의 표본의 신뢰구간 중 95개의 신뢰구간이 모수의 참 값을 포함하고 있음

예를 들어, 갑 후보의 지지율이 70%이고 표본오차가 3% / 신뢰수준이 90%라면 이는 90% 신뢰수준에서 신뢰구간이 67% ~ 73%라는 의미
