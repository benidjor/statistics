## 6.1 귀무(영)가설과 대립가설

### 6.1.1 귀무가설과 대립가설
- 가설 : 모집단에 대한 어떤 주장, 기술(description), 또는 가정
- 가설 검정 : 가설의 타당성 여부를 판단 <br>
<br>

- 귀무가설 (Null Hypothesis) $H_0$
- 대립가설 (Alternative Hypothesis) $H_1$
    - 대체로 새로운 주장, 흥미 대상이 되는 가설, 알고 있던 것과 다른 사실 등을 보통 $H_1$로 설정한다.

### 6.1.2 양측 검정과 단측 검정
- 양측 검정 (two-sided test) : $=, \ne$가 포함된 가설 검정 (달라졌다고 주장할 수 있는가)<br>

    $$H_0 : P(앞면) = 0.5 \\
    H_1 : P(앞면) \neq 0.5$$


- 단측 검정 (one-sided test) : 크다, 작다의 방향이 결정된 가설 검정<br>

    $$H_0 : \mu \leq 10 (또는 \mu = 10) \\
          H_1 : \mu > 10$$



## 6.2 검정통계량과 귀무가설의 기각역

- 검정통계량 (test statistic) : 검정의 목적으로 사용되는 통계량
    - 모집단 평균 ($\mu$)에 대한 가설이므로 표본으로 부터 얻는 표본평균 ($\bar{X}$)을 검정 통계량으로 잡아야 한다

- 귀무가설의 기각역 (rejection region of $H_0$)
    - 검정통계량의 구체적인 값이 어느 범위에 있으면 귀무가설을 기각할 것인가
    - 다시말해, 대립가설이 뒷받침되기 위해서는 검정통계량의 값이 어디에 있어야 하는지 정하는 범위

- 기각역 예시) 동전을 100번 던질때 앞면이 50번 나올 것으로 기대됨
$$ 검정통계량 : X = 앞면의 횟수 (100번 중) \\
H_0 기각역 : X < 50 - a 또는 X > 50 + a$$

- $H_0$의 기각역
    - 구체적인 표본 값으로 계산된 검정통계량의 값이 기각역 범위에 있다 
        - → $H_0$을 기각하고 대립가설 ($H_1$)의 내용을 받아들일 수 있다는 의미
- 통계학의 이론은 합리적 사고를 할 때 얻어지는 것을 논리적으로 표현하여 놓은 것이다

## 6.3 유의수준

### 6.3.1 제 1종 오류 (Type 1 error)
- 동전 던져서 앞면 나올 확률이 0.5라고 했는데 100번 던졌더니 21번 앞면 나왔다
    - 이때, 귀무가설 ($H_0$ = 0.5)를 기각했다면 오류를 범한 것 → 제 1종 오류 (Type 1 error)
    - 정상적인 동전이라도 100번 중 앞면이 21번 나올 수 있기 때문 <br>


- 유의 수준 (significance level): 1종 오류를 범할 확률
$$\alpha = P(제 1종 오류를 범함) = P(H_0 기각 | H_0)$$
- 일반적으로 $\alpha = 0.05(5\%) 사용$

- 귀무가설이 맞다 ($H_0$ 참)는 가정하게 검정통계량 $\bar{X}$의 분포는
$$ \bar{X} \sim N(\mu_0, \frac{\sigma^2}n)$$

## 6.4 가설검정의 예

### 6.4.1 $\mu$에 대한 가설검정

- 모집단 평균에 대한 가설 검정은 가장 자주 다루게 되는 것일뿐 아니라 가설 검정 과정을 이해하는 기초가 된다
    - 모집단 평균이 특정한 값 ($\mu_0$)이라고 할 수 있는가 없는가?
    $$ H_0 : \mu = \mu_0 vs H_1 : \mu \ne \mu_0 $$
    - 이 때, 사용하는 검정통계량 = $\bar{X}$의 분포
    $$ \bar{X} \sim N(\mu_0, \frac{\sigma^2}n)$$
    - 이를 표준화하면,
    $$ Z = \frac{\bar{X}-\mu}{\sigma/\sqrt{n}} $$
    - 그런데 $\sigma$를 모르기 때문에 $\sigma$ 대신 $S$를 사용하면 자유도 $(n-1)$인 $t-분포$를 따른다
    $$ T = \frac{\bar{X}-\mu}{S/\sqrt{n}}$$

    - 이 때, 앞의 유의수준 ($\alpha$)을 정의한 식
    $$
    \begin{align}
    \alpha &= P[H_0 기각 | H_0 참] \\
    &= P[H_0 기각\ |\ \mu=\mu_0]
    \end{align}
    $$
    - 으로 부터 $H_0$을 채택할 확률은
    $$ 1-\alpha = P[H_0 채택 | \mu = \mu_0] $$

    

    - (1-$\alpha$) : 신뢰수준
    - 표본을 통해 계산된 표본 평균값이 기각역에 있다면, 모집단 평균이 $\mu_0$가 아니라고 판단할 수 밖에 없다
        - 이러한 오류 (제 1종 오류 : $H_0$가 맞는데 $H_0$를 기각한다)를 범할 확률이 $\alpha$ (유의수준) 이다


### 6.4.2 비율 ($p$)에 대한 가설검정

- 예를 들어, 어떤 기업의 시장 점유율이 기존의 시장 점유율 $p_0$을 기준으로 낮아졌는지 알아본다
$$ H_0 : p = p_0 \quad vs \quad  H_1 : p < P_0 $$

- 비율에 대한 가설 검정의 검정 통계량 : 표본비율 $\hat{p}$
- 표본 비율은 중심극한 정리에 의하여 정규 분포로 간주할 수 있다
$$\hat{p} \sim N(p, \frac{p(1-p)}n)$$

- 평균에 대한 가설검정과 마찬가지로 기각역과 유의수준을 설정할 수 있다

## 6.5 유의확률 $p$ 값

- 일반적으로 $p$ 값이란 실제 계산된 검정통계량의 값으로부터 꼬리 부분의 면적이라고 이해하면 된다
- $p$ 값은 어떤 가설검정의 문제인가에 따라 문제에 맞는 특정한 방법으로 계산됨
- 예를 들어 프로그램에서는, 평균에 대한 가설검정을 양측검정 문제로 다룬다고 가정한다
$$ H_0 : \mu = \mu_0 \quad vs \quad H_1 : \mu \neq \mu_0 $$
$$
\begin{align}
p 값 &= 2P[\bar{X} > \bar{x}] \\
&= 2{[\frac{\bar{X} - \mu_o}{S/\sqrt{n}} > \frac{\bar{x} - \mu_o}{S/\sqrt{n}}]} \\
&= 2P[T>t]
\end{align}
$$

- 이럻게 되면, $\bar{x}$로 부터 꼬리 부분의 면적을 2배 한 값으로 $p$ 값이 계산 된다
- 유의 수준 $\alpha$ 과 $p$ 값은 밀접한 관계가 있다
- 계산된 $p$ 값이 매우 작다면, 입력된 자료의 표본 평균 값 $\bar{x}$가 $H_0$의 기각역에 있다는 것
    - 따라서 $H_0$을 기각하면 된다

- $p$ 값 : 유의확률 (significant probability)
    - 계산된 $p$값이 작을 경우 (아래 3가지 중 하나로 말하면 된다)
        - $p$ 값 수준에서 $H_0$을 기각한다고 하면 된다.
        - $p$ 값 수준에서 유의적이다
        - $p$ 값 수준에서 $H_1$의 내용이 뒷받침된다

- 유의확률 $p$값은 실제로 얻은 표본값에서 계산된 제 1종 오류를 범할 확률
    - $H_0 : \mu = \mu_0$가 참인데도 표본 평균 값이 \mu_0과 차이가 나서 \mu_0를 기각하는 위험의 정도를 나타낸다

## 6.6 분산 $\sigma^2$에 대한 가설검정

- 분산이라는 개념은 위험 또는 안정성을 측정하는 데 사용되기 떄문에 중요하다

- 모집단의 분산이 특정한 값 $\sigma_0^2$ 인가?
$$ H_0 : \sigma = \sigma_0^2 \quad vs \quad H_1 : \sigma \neq \sigma_0^2 $$

- 이 때, 검정 통계량으로 $S^2$ 를 사용한다
    - $S^2$가 $\sigma^2$에 대한 가장 좋은 추정량이기 때문

- $H_0$이 맞다는 전제하게서 $S^2$의 분포를 사용하여 가설을 검정한다
$$\frac{(n-1)S^2}{\sigma_0^2} \sim \chi^2_(n-1) $$

- 카이제곱 분포와 $H_0$의 기각역 (양측검정)



## 6.7 두 모집단 간의 차에 대한 검정

- 두 개의 서로 독립인 모집단의 평균값 $\mu_1, \mu_2$,
- 두 개의 평균값들에 차이가 있는가?

$$ H_0 : \mu_1 = \mu_2 \leftrightarrow \mu_1 - \mu_2 = 0 \\
H_0 : \mu_1 \neq \mu_2 \leftrightarrow \mu_1 - \mu_2 \neq 0

- 이럴 때, 두 모집단의 분산이 같은 경우 / 분산이 다를 경우로 나눠서 생각해야 한다
$$ (A) H_0 : \sigma_1^2 = \sigma_2^2 \\
(B) H_1 : \sigma_1^2 \neq \sigma_2^2

### 6.7.1 $\sigma_1^2 = \sigma_2^2 = \sigma^2$인 경우

- 서로 독립인 두 집단의 평균 차에 대한 검정을 위해, $\sigma_1^2 = \sigma_2^2$의 가정이 타당한지를 먼저 검정

    - 타당한 경우에는, 가설 검정을 한다
    
    - 즉, $\sigma_1^2 = \sigma_2^2$ 일 때 두 개의 모집단 분산이 갖기 때문에 두 표본을 합쳐 (pool) $\sigma^2$를 추정
    - 이는 자유도 $(n_1 + n_2 - 2)의 t-분포$를 따르게 된다

### 6.7.2 $\sigma_1^2 \neq \sigma_2^2 $인 경우

- 이러한 경우, $\sigma_1^2, \sigma_2^2$ 를 사용할 수 없다
    
    - 각각 ${X_1, X_2, ... ,X_n}, {Y_1, Y_2, ..., Y_n}$ 에서 추정한 값 $S_1^2$ 와 $S_2^2$ 를 사용할 수 밖에 없다

```python
# 예제 6-6
import numpy as np
import pandas as pd
```

```python
x = np.array([2.0, 2.0, 2.3, 2.1, 2.4, 2.2, 1.9, 2.5, 2.3, 2.4]) # x: 철 함유량
g = np.repeat(np.array(["A", "B"]), 5) # g: 분석 방법
d = {"g":g, "x":x}
data = pd.DataFrame(data=d)
data
```

```python
A = data[data.g=="A"]   # 집단 구분 A: 화학적
B = data[data.g=="B"]   # 집단 구분 B: X선

# 상황에 따른 기술통계량
data.groupby("g").x.describe()
```

```python
# 정규성 검정
from scipy.stats import shapiro
shapiro(x)
```

```python
# 등분산 검정 - Bartlett 검정 (정규성을 만족할 때)
from scipy import stats
stats.bartlett(A.x, B.x)
```

```python
# 등분산 검정 - Levene 검정 (정규성을 만족하지 않을 떄)
stats.levene(A.x, B.x)
```

```python
# T 검정 - 단측 검정, 이분산 가정
ttest_ind(A.x, B.x, alternative="smaller", usevar="unequal")
```

```python
# T 검정 - 양측검정, 이분산 가정
ttest_ind(A.x, B.x, alternative="two-sided", usevar="unequal")
```

## 6.8 짝으로 된 자료에 대한 검정

- 서로 관련이 있는 두 집단 간의 차이를 검정하는 경우

- 짝 (pair)으로 얻어지는 자료 : 하나의 개체에 대한 두 가지 자료
    - 모집단이 하나인 경우와 같다

- 두 가지 자료의 평균이 같은지 검정하는 가설
$$ H_0 : \mu_D = 0 \quad vs \quad H_1 : \mu_D \neq 0 $$

- 가설 검정하고자 하는 문제를 분명하게 구별해야 한다

    - 서로 독립인 두 집단의 비교인지 / 한 보집단의 두 가지 자료 집단의 비교인지 

```python
# 예제 6.7
import numpy as np
import pandas as pd
from scipy.stats import ttest_rel

x = np.array([224, 270, 400, 444, 590, 660, 1400, 680])
y = np.array([116, 96, 239, 329, 437, 597, 689, 576])
d = {"y":y, "x":x}
data = pd.DataFrame(data=d)

# 검정하고자 하는 가설 : 저장 후 소르빈산 잔여량이 감소하였는가? (대립가설, 단측검정)
# H_0 : \mu_D <= 0 
# H_1 : \mu_D > 0 (책에 적힌 내용, 잘못된 것 아닌가?)
```

```python
#  대응 T 검정 - 양측 검정
ttest_rel(x, y)
```

```python
# 대응 T 검정 - 단측검정
# p-value < 0.05, H_0 기각, 저장한 햄의 소르빈산 잔여량은 저장 전보다 감소하였다고 볼 수 있다
pval = ttest_rel(x, y)[1]
print("one-sided p-value =", pval/2)
```

```python
# 예제 6-8
# 금융업 신입사원의 근속일수는 1950일 이라고 주장할 수 있을까?
# H_0 : \mu = 1950 
# H_1 : \mu != 1950

import numpy as np
x = np.array([2000, 1975, 1900, 2000, 1950, 1850, 1950, 2100, 1975])

# 일표본 T 검정
import scipy.stats as stats
stats.ttest_1samp(x, popmean=1950)
# popmean : 모평균
# 예제의 모평균이 1950인지 가설검정하는 문제이기 때문에 popmean의 값은 1950

# p-value > 0.05, 이 금융업의 신입사원 근속 일수는 1950일이라고 주장할 수 있다
```

## 6.9 두 모집단 비율들의 차에 대한 검정

- 하나의 모집단에서 비율의 문제는 베르누이 분포를 따르는 변수 $X$에 대한 평균 $p$의 문제이다

- 여기서는 두 개의 집단에 대한 비율간의 차이에 대한 문제를 다루고자 한다

$$ X \sim Bernoulli \\
Y \sim Bernoulli $$

$$ H_0 : p_1 = p_2 \rightarrow p_1 - p_2 = 0 \\
H_0 : p_1 \neq p_2 \rightarrow p_1 - p_2 = 0 $$


- 추정량은 귀무가설이 옳다고 가정하고 표본을 합한 (pool) 경우를 이용하여, 두 모집단의 비율이 같다고 가정한 통합표본비율 $\hat{p}$ 사용한다
$$ \hat{p} = \frac{\sum{X_i} + \sum{Y_i}}{n_1 + n_2} $$

- 따라서, $\hat(p_1) - \hat(p_2)$의 표준화 변수 $Z$는
$$ Z = \frac{((\hat{p_1}-\hat{p_2}) - (p_1-p_2))}{\sqrt{\hat{p}(1-\hat{p})(\frac{1}{n_1}+\frac{1}{n_2})}}

```python
# 예제 6.9
import pandas as pd
data = pd.DataFrame([[56, 24], [44, 37]], index=["A", "B"], columns=["use", "unuse"])
data
```

```python
# A 지역의 시장점유율 p_A가 B 지역의 시장점유율 p_B보다 높은지를 알아보고자 한다
# H_0 : p_A <= p_B ↔ p_A - p_B <= 0
# H_1 : p_A > p_B ↔ p_A - p_B > 0

# 모비율 검정은 Fisher's exact test를 이용한다 : fisher_exact()

from scipy.stats import fisher_exact
fisher_exact(data, alternative="greater")
# p-value < 0.05 → H_0 기각
```
