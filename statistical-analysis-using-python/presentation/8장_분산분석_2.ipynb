## 8.6 잔차분석

- 잔차는 예측값 - 실제값(관측값)을 의미함. 즉, 예측값으로 설명되지 않는 부분
- 분산분석은 오차의 등분산성을 가정하므로 잔차의 분산이 일정해야 한다.
- 잔차의 분산은 다음과 같이 정의할 수 있다. 

$$S^2 = \frac{\sum (e_i-\bar{e})}{\alpha}$$

    - 이때, 잔차의 평균은 0이므로 이는 다시 다음과 같이 정의될 수 있다. 
    
    $$S^2 = \frac{\sum (e_i)}{\alpha}$$

- 그러나, 잔차는 관찰치의 측정 단위로 나타나므로 이를 표준화시켜야 한다. 이를 스튜던트화 잔차로 표현하며, 구체적인 식은 다음과 같다.
$$스튜던트화 잔차 = \frac{e_i}{(e_i의 표준편차)}$$

    - 이때, 스튜던트화 잔차값의 관찰치 중 -2 ~ 2를 넘어가는 범위는 이상관찰치라 하며 주목할 필요가 있다. 
    
- 또한, 잔차의 산점도를 그려서 잔차의 등분산성을 살펴볼 수도 있다. 아래 그림 중, 잔차가 점점 퍼지는 모양은 잔차가 균등하게 분포되어 있지 않다는 것을 의미한다.

## 8.7 자료변환

- 분산분석은 일반적으로 서로 다른 처리에서 같은 분산을 가지고 있다고 가정한다. 그러나, 많은 경우 서로 다른 처리에서 서로 다른 분산을 갖는 경우도 많다. 
- 이러한 상황에 8.11에서 다루는 하틀리 검정 방법을 사용할 수도 있지만, 원자료를 함수 변환하여 사용할 수도 있다. 
- 즉, 잔차가 일정한 범위 안에서 분포하지 않고 특별한 패턴을 보일 경우, 자료의 특성에 따라 간단한 함수 관계로 변환시키는 것이다. 

### 1) 제곱근변환
- 제곱근변환은 각 처리에서 관찰치의 평균이 커질수록 관찰치 분산이 커지는 자료에 유용하다. 즉, 일차식의 잔차 분포를 보이는 경우에 사용한다. 변환식은 아래와 같으며 0.375는 고정값이다.

$$\sqrt{y+0.375}$$

### 2) 대수변환
- 대수변환은 관찰치들 분산이 관찰치의 평균의 증가보다 현저하게 증가할 때 원자료에 대수(로그)를 취한다. 즉, 제곱식 형태의 잔차 분포를 보이는 경우에 사용한다.

### 3) 제곱변환
- 제곱근변환의 역관계에 사용할 수 있다.

### 4) 역사인 함수변환
- 관찰치가 비율로 얻어진 경우 사용하는 변환으로 식은 아래와 같다. 

$$\arcsin(0.5) = 30$$

## 8.8 비모수적 분산분석 : 순위자료

- 모집단이 정규분포라 가정하지 못할 경우 사용하는 분석으로 정규성이 아닌 연속형/대칭과 같은 간단한 가정만을 세우고 진행하는 분산분석을 의미한다.

### 1) 크루스칼-왈리스 검정
- 일원배치분산분석에 대응하는 비모수통계적 방법이다.
- 방법을 간략히 소개하면, 처리가 이루어진 집단별로 전체 순위의 평균을 구한다. 예를 들어 집단 '갑'의 순위가 1,3,5,7,9이고 집단 '을'의 순위가 2,4,6,8,10이면 갑의 순위평균은 5, 을의 순위평균은 6이된다. 
- 이렇게 해서 구한 순위평균의 평균(모든 순위평균)을 구한다. 
- 이때, 모든 순위평균은 1부터 n까지의 합을 개수로 나눈 것과 같다. 예를 들어, 10개가 있으면 1부터 10까지 더한 값인 55를 10으로 나눈 것과 같다. 

$$\bar{R} = \frac{1}{n}\sum\limits_{i=1}^k \sum\limits_{j=1}^{n_i} R_{ij} = \frac{1}{n}(\frac{1}{2}n(n+1)) = \frac{1}{2}(n+1)$$

- 이렇게 하여 구한 검정통계량은 다음 식과 같으며 자유도가 k-1인 카이제곱을 따른다. 이를 미리 구해진 값과 비교하는 방식을 사용한다. 

$$H = \sum\limits_{i=1}^k \sum\limits_{j=1}^{n_i}(\bar{R_i}-\frac{n+1}{2})^{2}/S^{2}$$

- 단, 이때 통상적인 분산분석과 달리 크루스칼-왈리스 검정의 귀무가설과 대립가설은 다소 다르게 표현한다. 그러나 그 의미는 같다.
    - 귀무가설 : k개 모집단의 분포 형태는 일치한다.
    - 대립가설 : 적어도 하나의 모집단은 다른 하나의 모집단보다 큰 관찰치를 포함하고 있다.

## 공분산분석
- 공분산분석이란 분산분석과 회귀분석을 혼합한 형태의 분산분석이다.
- 즉 분산분석처럼 독립적인 k개 모집단 평균(처리효과) 간에 차이를 검정하지만, 자료(Y)가 다른 변수와 어떤 관계에 있다고 믿어질 때 유용한다.
- 예를 들어, 화장품 대리점의 판매가 해당 지역의 생활 수준에 따라 다른가를 분산분석할 때, 매장의 크기를 함께 반영하는 경우에 사용한다.
- 식은 다음과 같다. 이때, mu는 전체 평균, alpha는 처리집단 평균, beta는 회귀계수, x는 관측치, 입실론은 잔차를 의미한다. 

$$Y_{ij} = \mu+\alpha_i+\beta x_{ij}+\epsilon_{ij}$$

## 8.10 두 진단 분산 간의 차에 대한 가설검정
- 잠시 분산분석이 아니라 두 개 모집단의 분산 차이 검정으로 돌아가보자. 
- 두 개 모집단의 분산 차이 검정은 두 개 모집단으로부터 추출된 표본들의 분산을 구하고, 이 비교값을 구하여 가설을 검정하게 된다.
- 이때, "두 개 모집단으로부터 추출된 표본들의 분산의 비율"이 바로 F통계량의 정의이므로, 두 개 모집단의 분산 차이 검정은 F통계량을 이용하여 검정한다.
